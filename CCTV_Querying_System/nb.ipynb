```python
from langchain_ollama import OllamaEmbeddings
from langchain_chroma import Chroma

# Define ChromaDB directory
CHROMA_DB_DIR = "./chroma_db"

def read_chroma_db():
    """Reads and displays data stored in ChromaDB."""
    try:
        # Initialize embeddings (must match the model used when storing data)
        embeddings = OllamaEmbeddings(model="nomic-embed-text")
        
        # Load the existing ChromaDB
        vector_store = Chroma(embedding_function=embeddings, persist_directory=CHROMA_DB_DIR)
        
        # Fetch all stored documents
        results = vector_store._collection.get(include=["metadatas", "documents"])
        
        # Display results
        for idx, doc in enumerate(results["documents"]):
            metadata = results["metadatas"][idx]
            print(f"Document {idx + 1}:")
            print(f"Metadata: {metadata}")
            print(f"Content: {doc}\n")
    
    except Exception as e:
        print(f"Error reading ChromaDB: {str(e)}")

# Call the function
read_chroma_db()
```
